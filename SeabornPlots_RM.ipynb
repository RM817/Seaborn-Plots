```python
#Assignment 1 (Part 1): Players salary is the variable and it is plotted using bar, strip, stem, and horizontal bar plots.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv("NBA_season1718_salary.csv") 
sns.set(style="darkgrid")
data.season17_18 = data.season17_18/1000000

g = sns.catplot(x="Player", y="season17_18", hue="Tm", data=data.head(),
                   height=6,kind='bar',aspect = 2)

g.set_ylabels("Salary (million)")
g.set_xlabels("")
plt.title('Barplot: Top Player Salaries')
plt.show()

#Pros: Easy to understand the important factors within the graph.

#Cons: Hard to spot small differences and quantify them.
```

```python
g2 = sns.catplot(x="Player", y="season17_18", hue="Tm", data=data.head(),
                   height=6,kind='strip')

g2.set_ylabels("Salary (million)")
g2.set_xlabels("")
plt.title('Stripplot: Top Player Salaries')
plt.show()

#Pros: Easy to understand the differences in pay for each player.

#Cons: Y-Axis can lead to misconceptions.
```

```python
g3 = plt.stem(data["Player"].head(), data["season17_18"].head(),use_line_collection = True)
plt.title('Stem Plot: Top Player Salaries')
plt.show()

#Pros: A condensed version of a bar graph that can be good for evolution a stat over time.

#Cons: Can get noisy with large amounts of data.
```

```python
g4 = sns.catplot(x="season17_18",y="Player",data=data.head(), height = 6, kind = 'bar')

g4.set_xlabels("Salary (million)")
g4.set_ylabels("")
plt.title('Horizontal Bar: Top Player Salaries')
plt.show()

#Pros: Easy to understand and retrieve important information quickly.

#Cons: Vertical bars are typically preferred over horizontal.
```

```python
#Assignment 1 (Part 2): Draft position and Years played is plotted using stem, scatter, and heatmap plots.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data2 = pd.read_csv("draft78.csv")
delete_row = data2[data2["Pick"]>60].index
data2 = data2.drop(delete_row)
data2 = data2.fillna(0)
data2 = data2.drop(columns = ['Player','Draft'])
draftlist = set(data2.Pick)
grouped = data2.groupby('Pick')
data2 = grouped.aggregate(np.average)
data2['Pick Order'] = draftlist

g = plt.stem(data2["Pick Order"], data2["Yrs"],use_line_collection = True)
plt.title('Stem Plot: Average Years Played Per Pick')
plt.show()
#Pros: Easily spot differences as you increase the pick value.

#Cons: Hard to individually pinpoint each pick's value. Is noisy with large amounts of data.
```
